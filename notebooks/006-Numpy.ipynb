```python
class Polynomial:
    
    def __init__(self, *coefs):
        self.coefs = coefs
        
    def _repr_latex_(self):
        s = [(f"{a}x^{i}" if a else "") for i, a in enumerate(self.coefs)] 
        return f'$${"+".join(s)}$$'
```

```python
Polynomial(3, 2, 0)
```

```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np
```

```python
t = np.arange(-3, 3, 0.01)
s = 1 + np.cos(2*np.pi*t)
plt.plot(t, s, '.-g')
plt.plot(t, t*t, '-r')
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('About as simple as it gets, folks')
plt.grid(True)
```

```python
import json

with open('us-cities.json') as f:
    cities = json.load(f)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

xy = [(city['longitude'], city['latitude']) for city in cities]
plt.scatter(*zip(*xy))
plt.grid('on')
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

xy = [(city['longitude'], city['latitude']) for city in cities]
size = [int(city['population'])/10000  for city in cities]
plt.scatter(*zip(*xy), s=size, alpha=0.4)
plt.axis('off');
```

```python
x, y = np.meshgrid(np.r_[-1:1:0.01], np.r_[-1:1:0.01])

fig, ax = plt.subplots(1, 3, figsize=(10, 4))
ax[0].imshow(x)
ax[1].imshow(y)
ax[2].imshow(x*x+y*y)
```

```python
x, y = np.meshgrid(np.r_[-1:1:0.01], np.r_[-1:1:0.01])
sigma = 0.1
fig, ax = plt.subplots(1, 3, figsize=(10, 4))
ax[0].imshow(x)
ax[0].axis(False)

ax[1].imshow(y)
ax[1].axis(False)

ax[2].imshow(np.exp(-(x*x+y*y)/2/sigma/sigma))
ax[2].axis(False)
```

```python
x, y = np.meshgrid(np.r_[-1:1:0.01], np.r_[-1:1:0.01])
sigma = 0.1
fig, ax = plt.subplots(1, 3, figsize=(10, 4))

ax[0].imshow(x)
ax[0].axis(False)

ax[1].imshow(y, cmap='cubehelix')
ax[1].axis(False)

cx, cy = 0, -0.3 
z = np.exp(-((x-cx)**2+(y-cy)**2)/2/sigma/sigma)
cx, cy = 0.2, 0.3 
z -= np.exp(-((x-cx)**2+(y-cy)**2)/2/sigma/sigma)

ax[2].imshow(z, cmap='Spectral')
ax[2].axis(False)
```
