```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
x, y = np.meshgrid(np.linspace(-1, 1, 300), np.linspace(-1, 1, 300))
c = x + 1j*y

z = np.zeros_like(c)
for i in range(10):
    z = z*z + c

plt.imshow(abs(z) > 2, cmap='gray')
plt.axis('equal')
plt.axis(False)
```

```python
x, y = np.meshgrid(np.linspace(-.50, -.55, 1000), np.linspace(-.50, -.55, 1000))
c = x + 1j*y
im = np.zeros_like(c, dtype=np.float64)
z = np.zeros_like(c)

for i in range(120):
    ix = im == 0
    z[ix] = z[ix] * z[ix] + c[ix]
    im[np.logical_and(abs(z) > 2, ix)] = i
```

```python
fig, ax = plt.subplots(1, 1, figsize=(24, 24))
ax.imshow(im, cmap='inferno')
ax.grid(True)
```
